### Soft K Means: An Introduction


Let us briefly go through what is the K means algorithm. Let us a consider a data spanning over 2D space of two features $x$ and $y$. When $x$ and $y$ are plotted against each other the data forms cluster and we would like our algorithm to identify these clusters on its own. Let us first define the data that needs to be analysed.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
D = 2
s = 5
mu1 = np.array([0, 0])
mu2 = np.array([s, s])
mu3 = np.array([0, s])

N = 3000 # Total points
X = np.zeros([N, D])
X[:1000] = mu1 + np.random.randn(1000, D)
X[1000:2000] = mu2 + np.random.randn(1000, D)
X[2000: N] = mu3 + np.random.randn(1000, D)
```

```python
plt.scatter(X[:,0], X[:, 1])
plt.tight_layout()
```

The plot of the $x$ vs $y$ data shows some sort of clustering that is happening. This cluster is almost trivial as this can be visually recognized. It will be very difficult to recognize such clusters by the addition of another dimension (or feature in this case). However the addition of another dimension should not be a problem for our algorithm as it should be able to take care of the clustering irrespective of the number of dimensions, albeit with additional computational time.
